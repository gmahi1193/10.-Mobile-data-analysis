## Create a new column 'payment_type' and categorize all the applications into 'free' or 'paid'

```python
# We will be using the apply function here to get results

# Custom function to classify an app as free or paid
def is_free(x):
    """
    Checks if the app paid or free
    
    Params:
    -------
    price: int or float
    
    Returns:
    --------
    free if price is 0 else paid
    """
        
    if x > 0.00:
        return 'paid'
    else:
        return 'free'

# creating a new column as per our requirement
apple['payment_type'] = apple['price'].apply(lambda x : is_free(x))

apple.head()
```

### Create a new column "size_mb" converting bytes to megabytes. For the applications below 50 dollars, does the price of the app increases with the increase in its size?

```python
# Converting bytes to megabytes
def size(x):
    mb = x/1000000
    return mb
    
top_genre_df['size_mb'] = apple.size_bytes.apply(lambda x : size(x))
top_genre_df.head(5)
```

```python
# creating a scatter plot to get our answer
plt.figure(figsize = (15,10))
paid_apps_desc = top_genre_df[top_genre_df['price'] <50]

sns.lmplot(data = paid_apps_desc, x = 'size_mb', y = 'price', col = 'prime_genre',col_wrap= 3,aspect= 1.5
           ,scatter = True,fit_reg = False, hue  = 'prime_genre', legend_out = True,
           palette = ['#c973d0' , '#8e52b8' , '#4a4f92', '#4a73ab' , '#649ca1'] )
plt.show()
```

* The relation between Price and Genere suggest that there exists no relation between app size and price

## Can we create a recommendation column based on "user_rating" and "rating_count_tot"?

```python
def recommendation(total_reviews,rating):
# creating an empty list to store the recommendation
    recommendation = []
# Iterating through all the rating and review count values, using the if else statement and categorizing the app
    for review_count,avg_rating in zip(apple['rating_count_tot'],apple['user_rating']):
        if review_count > total_reviews and avg_rating > rating:
            recommendation.append('yes')
        else:
            recommendation.append('no')
# Creating a new recommendation column
    apple['recommendation'] = recommendation
```

```python
recommendation(20000,3.5)
```
